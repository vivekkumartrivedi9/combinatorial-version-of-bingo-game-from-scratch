## strategy

```python
def ge(a,n):
  d=[]
  for i in range(n):
    count=0
    for j in range(n):
      if a[i][j]==0:
        count+=1
    d.append(count)
  for i in range(n):
    count1=0
    for j in range(n):
      if a[j][i]==0:
        count1+=1
    d.append(count1)
  counta=0
  for i in range(n):
    if a[i][i]==0:
      counta+=1
  d.append(counta)
  countb=0
  for i in range(n):
    if a[i][n-1-i]==0:                                                                                                                                                                                                                                                                                                                                                               
      countb+=1
  d.append(countb)
  for j,k in enumerate(d):
    if k==n:
      d[j]=0
  e=max(d)
  for j,k in enumerate(d):
    if k==e:
      if j<n:
        #row
        vr=[]#position
        cou=[]
        for i in range(len(a[j])):
          if a[j][i]!=0:
            couy=0
            vr.append(i)
            for v in range(n):
              if a[v][i]==0:
                couy+=1
            cou.append(couy)
        ou=max(cou)
        for i in range(len(cou)):
          if cou[i]==ou:
            return a[j][vr[i]]
      if n<=j<2*n:
        vi=[]
        for v in range(n):
          vi.append(a[v][j-n])
          #coloumn 
        vr=[]
        cou=[]
        for i in range(len(vi)):
          if vi[i]!=0:
            couy=0
            vr.append(i)
            for v in range(n):
              if a[i][v]==0:
                couy+=1
            cou.append(couy)
        ou=max(cou)
        for i in range(len(cou)):
          if cou[i]==ou:
            return a[vr[i]][j-n]
      if j==2*n:
        vi=[]
        for v in range(n):
          vi.append(a[v][v]) 
        for i in vi:
          if i!=0:
            return i
      if j==2*n+1:
        vi=[]
        for v in range(n):
          vi.append(a[v][n-1-v]) 
        for i in vi:
          if i!=0:
            return i
```

##checking for winners

```python
def ch(a,c,n,m):
  count=0
  count1=0
  for i in range(n):
    co=0
    for j in range(n):
      if a[i][j]==0:
        co+=1
    if co==n:
      count+=1
    co=0
    for j in range(n):
      if a[j][i]==0:
        co+=1
    if co==n:
      count+=1
  co=0
  for i in range(n):
    if a[i][i]==0:
      co+=1
  if co==n:
    count+=1
  co=0
  for i in range(n):
    if a[i][n-1-i]==0:
      co+=1
  if co==n:
    count+=1  
  for i in range(n):
    co=0
    for j in range(n):
      if c[i][j]==0:
        co+=1
    if co==n:
      count1+=1
    co=0
    for j in range(n):
      if c[j][i]==0:
        co+=1
    if co==n:
      count1+=1
  co=0
  for i in range(n):
    if c[i][i]==0:
      co+=1
  if co==n:
    count1+=1
  co=0
  for i in range(n):
    if c[i][n-1-i]==0:
      co+=1
  if co==n:
    count1+=1   
  if count>=m:
    return 'first player won'
  if count1>=m:
    return 'computer won' 
```
